# Customer Table Analysis

```python
import pandas as pd
import numpy as np

customers = pd.read_csv("blinkit_customers.csv")

# Basic Information
customers.info()
customers.describe(include="all")

# Check Shape
customers.shape

# Missing Values
customers.isnull().sum()

# Duplicate Rows
customers.duplicated().sum()

# Duplicate Customer IDs
customers["customer_id"].duplicated().sum()

# Remove Duplicate Rows
customers.drop_duplicates(inplace=True)

# Registration Date
customers["registration_date"] = pd.to_datetime(
    customers["registration_date"],
    errors="coerce"
)

# Fill Missing Customer Segment
customers["customer_segment"] = customers["customer_segment"].fillna("Unknown")

# Total Orders
customers["total_orders"] = customers["total_orders"].fillna(0)

# Average Order Value
customers["avg_order_value"] = customers["avg_order_value"].fillna(
    customers["avg_order_value"].median()
)

# Pincode should be String
customers["pincode"] = customers["pincode"].astype(str)

# Feature Engineering
customers["registration_year"] = customers["registration_date"].dt.year
customers["registration_month"] = customers["registration_date"].dt.month_name()
customers["phone"] = customers["phone"].astype(str)
customers["customer_segment"] = customers["customer_segment"].astype("category")
month_order = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]

customers["registration_month"] = pd.Categorical(
    customers["registration_month"],
    categories=month_order,
    ordered=True
)
customers.head()
```

```python
customers.info()
```

```python
customers.isnull().sum()
```

```python
customers.to_csv(
    "cleaned_customers.csv",
    index=False
)
```

# Data Integration

```python
# ================================
# Customer Analysis Dataset
# ================================

customer_analysis = (
    customers
    .merge(orders, on="customer_id", how="left")
    .merge(
        feedback[
            ["order_id","rating","sentiment","feedback_category"]
        ],
        on="order_id",
        how="left"
    )
)

# ================================
# Sales Analysis Dataset
# ================================

sales_analysis = (
    orders
    .merge(order_items, on="order_id", how="left")
    .merge(products, on="product_id", how="left")
    .merge(
        customers[
            ["customer_id","customer_name","customer_segment"]
        ],
        on="customer_id",
        how="left"
    )
)

# ================================
# Product Analysis Dataset
# ================================

product_analysis = (
    order_items
    .merge(products, on="product_id", how="left")
)

# ================================
# Inventory Analysis Dataset
# ================================

inventory_analysis = (
    inventory
    .merge(products, on="product_id", how="left")
)

# ================================
# Delivery Analysis Dataset
# ================================

delivery_analysis = (
    delivery
    .merge(orders, on="order_id", how="left")
    .merge(
        feedback[
            ["order_id","rating","sentiment","feedback_category"]
        ],
        on="order_id",
        how="left"
    )
)

# ================================
# Marketing Analysis Dataset
# ================================

marketing_analysis = marketing.copy()

# ================================
# Master Dataset
# ================================

master = (
    orders
    .merge(customers, on="customer_id", how="left")
    .merge(order_items, on="order_id", how="left")
    .merge(products, on="product_id", how="left")
    .merge(delivery, on="order_id", how="left")
    .merge(feedback, on=["order_id","customer_id"], how="left")
)
```

# Export Final Clean Data

```python
master.to_csv("master_dataset.csv", index=False)

customer_analysis.to_csv("customer_analysis.csv", index=False)

sales_analysis.to_csv("sales_analysis.csv", index=False)

product_analysis.to_csv("product_analysis.csv", index=False)

inventory_analysis.to_csv("inventory_analysis.csv", index=False)

delivery_analysis.to_csv("delivery_analysis.csv", index=False)

marketing_analysis.to_csv("marketing_analysis.csv", index=False)
```
